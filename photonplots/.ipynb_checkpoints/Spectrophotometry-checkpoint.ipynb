```python
from speccounts import *
import matplotlib.pyplot as plt
import numpy as np

```

```python
def loadtext(file,wavelist=[],fluxlist=[]):
  f = open('' +file,'r') #open file
  for line in f:
        line = line.rstrip()
        column = line.split() # split line into a list 
        wavelength = float(column[0]) #putting data into named variable for readability
        flux = float(column[1])
        wavelist.append(float(wavelength))
        fluxlist.append(float(flux))
  f.close()
   #x+=1     
  return wavelist, fluxlist
```

```python
file="../spectra/SN2022hrs_peak_11fe_appended.dat"
```

```python
wavelength_spectra, flux_spectra = loadtext(file)
```

```python
counts_out=specin_countsout(wavelength_spectra, flux_spectra)
```

```python
import astropy.units as u
from astropy.table import Table
from dust_extinction.parameter_averages import CCM89, F99
from synphot import units, config
from synphot import SourceSpectrum,SpectralElement,Observation,ExtinctionModel1D
from synphot.models import BlackBodyNorm1D
from synphot.spectrum import BaseUnitlessSpectrum
from synphot.reddening import ExtinctionCurve
from astroquery.simbad import Simbad
from astroquery.mast import Observations
import astropy.visualization
```

```python
Rv = 3.1  
Ebv = 0.36
ext = CCM89(Rv=Rv)
fig, ax = plt.subplots()
ax.plot(wavelength_spectra, flux_spectra)
wavelength_spectra_microns=np.array(wavelength_spectra)*0.0001 * u.micron
ax.plot(wavelength_spectra, flux_spectra/ext.extinguish(wavelength_spectra_microns,Ebv=Ebv))
counts_out=specin_countsout(wavelength_spectra, flux_spectra/ext.extinguish(wavelength_spectra_microns,Ebv=Ebv))
```
